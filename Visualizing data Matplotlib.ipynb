```python
from pandas.plotting import andrews_curves
# Look pretty...
matplotlib.style.use('ggplot')
# If the above line throws an error, use plt.style.use('ggplot') instead

# Load up SKLearn's Iris Dataset into a Pandas Dataframe
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target_names'] = [data.target_names[i] for i in data.target]

# Andrews Curves Start Here:
plt.figure()
andrews_curves(df, 'target_names')
plt.show()
```

```python
wheat = pd.read_csv("C:/Users/maria/OneDrive - Scandizzo & Partner, LLC/Education/Python/EDX Courses/Data 210/Module3/Datasets/wheat.data",
                   index_col='id')
```

```python
area_perimeter_wheat = wheat[['area','perimeter']]
groove_asymmetry_wheat = wheat[['groove','asymmetry']]
```

```python
area_perimeter_wheat.plot.hist(alpha=0.75);
groove_asymmetry_wheat.plot.hist(alpha=0.75);
```

```python
wheat.plot.scatter(x='area',y='perimeter');
wheat.plot.scatter(x='groove', y = 'asymmetry');
wheat.plot.scatter(x='compactness', y= 'width');
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('area')
ax.set_ylabel('perimeter')
ax.set_zlabel('asymmetry')

ax.scatter(wheat['area'],wheat['perimeter'],wheat['asymmetry'], c='r', marker='.')
plt.show();
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('width')
ax.set_ylabel('groove')
ax.set_zlabel('length')

ax.scatter(wheat['width'],wheat['groove'],wheat['length'], c='g', marker='.')
plt.show();
```
