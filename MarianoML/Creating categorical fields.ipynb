## Modify categorical fields
------------------------------------------

```python
educacion = set(census['education'])
raza = set(census['race'])
clasificacion = set(census['classification'])
print(educacion)
print(raza)
print(clasificacion)
```

```python
edu_rank =['Preschool','1st-4th','5th-6th','7th-8th','9th','10th','11th','12th','HS-grad','Some-college',
           'Bachelors','Masters','Doctorate']
clasi_rank = ['<=50K', '>50K']
```

```python
census['education_cat'] = census['education'].astype("category", ordered=True,
  categories=edu_rank).cat.codes
census['classi_cat'] = census['classification'].astype("category", ordered=True,
  categories=clasi_rank).cat.codes
```

## Creating one hot categorical fields 
----------------------------------------------------

```python
census = pd.get_dummies(data= census, columns= ['race'])
```
