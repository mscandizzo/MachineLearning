```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
dfs = pd.read_html('http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2',
                  header=1, flavor = 'bs4')
```

```python
dfs = dfs[0]
```

```python
dfs.dropna(axis = 0 , thresh=4, inplace=True)
```

```python
dfs.drop('RK',axis =1, inplace = True)
```

```python
dfs = dfs.drop_duplicates(subset=['PLAYER','TEAM'])
```

```python
dfs.drop(11,axis=0,inplace=True)
```

```python
dfs.reset_index(drop=True)
```

```python
dfs.columns
```

```python
columnas = ['GP', 'G', 'A', 'PTS', '+/-', 'PIM', 'PTS/G', 'SOG',
       'PCT', 'GWG', 'G.1', 'A.1', 'G.2', 'A.2']
```

```python
for columna in columnas:
    dfs[columna] = pd.to_numeric(dfs[columna], errors='coerce')
```
